```python
import pandas as pd
import re
 
data = pd.read_csv('./ProcessedDataSet/ClusterdResult.csv')
data.head()
```

```python
data.columns
```

```python
row  = ''.join(data.Tags.values)
tags = re.split('<|>',row)
tags = [Tag for Tag in tags if len(Tag)>1]
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

masterString=(" ").join(tags)
wordcloud = WordCloud(width = 1000, height = 500).generate(masterString)
# plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig('static/WC.png', dpi = 300)
```
