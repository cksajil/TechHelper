```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
```

```python
sentances = ['like icecream', 'unlike icecream']
```

```python
newsentane= ['hate cream']
```

```python
vectorizer              =  TfidfVectorizer(stop_words="english")
X                       =  vectorizer.fit_transform(sentances)
words                   =  vectorizer.get_feature_names()
```

```python
words
```

```python
points = np.array((X.todense()))
```

```python
vectorizer.transform(newsentane).toarray()
```

```python
x = points.T[0,:]
y = points.T[1,:]
z = points.T[2,:]
```

```python
from matplotlib import pyplot
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D


fig = pyplot.figure()
ax = Axes3D(fig)
ax.scatter(x, y, z)
ax.set_xlabel('icecream')
ax.set_ylabel('like')
ax.set_zlabel('unlike')
ax.text(0.57973867,0.81480247,0.,'like icecream')
ax.text(0.57973867,0, 0.81480247,'unlike icecream')

pyplot.show()
```
